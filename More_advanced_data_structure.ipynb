##**HW5 เขียน Function คูณ matrix**

แล้ว test กับ matrix

* (2,3)*(3,2)
* (4,4)*(4,1)
* (2,2)*(2,2)

```
import numpy as np
```

```
mat_T = np.array([[1,4,2],[2,8,7]])
mat_T
#Matrix 2*3
```

```
mat_O = np.array([[4,2],[6,8],[2,1]])
mat_O
#Matrix 3*2
```

```
mat_E = np.array([[1,5,4,3],[5,2,1,4],[9,6,2,4],[1,3,2,4]])
mat_E
#Matrix 4*4
```

```
mat_I = np.array([[1],[4],[5],[7]])
mat_I
#Matrix 4*1
```

```
mat_P = np.array([[5,1],[4,2]])
mat_P
#Matrix 2*2 
```

```
mat_A = np.array([[1,8],[7,3]])
mat_A
```

**Create Function Multiply Matrix**

```
def mul_mat(T, O):
    E = [[0 for row in range(len(O[0]))] for col in range(len(T))]
    for i in range(len(T)):
        for j in range(len(O[0])):
            for k in range(len(O)):
                E[i][j] += T[i][k]*O[k][j]
    r = np.matrix(E)
    return r
```

```
mul_mat(mat_T,mat_O)
#(2,3)*(3,2)
```

```
mul_mat(mat_E,mat_I)
#(4,4)*(4,1)
```

```
mul_mat(mat_P,mat_A)
#(2,2)*(2,2)
```

**ตรวจสอบความถูกต้อง**

```
np.dot(mat_T,mat_O)
#(2,3)*(3,2)
```

```
np.dot(mat_E,mat_I)
#(4,4)*(4,1)
```

```
np.dot(mat_P,mat_A)
#(2,2)*(2,2)
```
